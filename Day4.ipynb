Filtering

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image, ImageFilter
kernel = np.ones((100,100),np.float32)/10000
dst = cv2.filter2D(im,-1,kernel)

plt.subplot(121),plt.imshow(im),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

# Introduction to computer vision

```python
import cv2
```

```python
def compare_image(image1,image2):
  plt.figure(figsize=(9,9))
  plt.subplot(1,2,1)
  plt.imshow(image1)
  plt.title('original')
  plt.axis('off')
  plt.subplot(1,2,2)
  plt.imshow(image2)
  plt.title('modifies')
  plt.axis('off')
  plt.tight_layout()
```

Edge detection

```python
url = "https://www.cnet.com/a/img/iJxo9AIxiXHqVoqm6nGISKtKwPI=/2020/08/18/b7168aea-9f7e-47bb-9f31-4cb8ad92fbc7/lg-note-20-ultra-5g-iphone-11-se-google-pixel-4a-lg-velvet-6133.jpg"
image = io.imread(url) 
```

```python
edge = cv2.Canny(image,100,200)
compare_image(image,edge)
```
